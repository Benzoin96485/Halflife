```python
def empty_str(s):
    if s.strip():
        return s
    else:
        return None

def reduce_col(val_list, target):
    if target in val_list:
        return 1 / len(val_list)
    else:
        return 0

def reduce_cols(df, old_col, reduce_dict):
    all_col = []
    for l in reduce_dict.values():
        for name in l:
            if name not in all_col:
                all_col.append(name)
    for col in all_col:
        df["is_" + col] = df[old_col].apply(lambda x: reduce_col(reduce_dict[x], col))
    return df

def lgHL(val):
    pattern = "\d+\.?\d*"
    match = re.search(pattern, val)
    if match:
        return np.log10(float(match.group()))
    else:
        print(val)
```

```python
df = pd.read_csv(data_path, header=4)
df["Chemical Structure"] = df["Chemical Structure"].apply(empty_str)
df["Concomitants"] = df["Concomitants"].apply(empty_str)
df['Normalized Value'] = df['Normalized Value'].apply(empty_str)
df = df[df["Chemical Structure"].notna()]
df = df[df["Concomitants"].isna()]
df = df[df['Normalized Value'].notna()]
df["lgHL"] = df['Normalized Value'].apply(lgHL)
df = df[['Drug', 'Chemical Structure', 'Study Group', 'Age', 'Dose',
       'Duration', 'Route', 'lgHL', 'SD']]
df["smiles"] = df["Chemical Structure"].apply(smi_cleaner)
df = reduce_cols(df, "Age", ages)
df = reduce_cols(df, "Route", routes)
```
